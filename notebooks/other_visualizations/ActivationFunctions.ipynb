## ReLU

```python
x = list(range(-4, 5))
x
```

```python
relu = lambda x: max(0, x)
y = [relu(el) for el in x]
y
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = sns.lineplot(x, y)
ax.set(title='ReLU', xlabel='x', ylabel='relu(x)');
plt.axvline(0, linestyle='--', color='gray');
```
