# Data leakage tests

```python
import pandas as pd

y = pd.read_parquet('../data/targets_data/clusters/Galson_2015a.parquet')

y.info()
y.head()
```

```python
valid_data = pd.read_parquet('../data/features_data/raw_subject_split/X_valid.parquet')

valid_data.info()
valid_data.head()
```

```python
valid_cdr3 = set(valid_data['cdr3'])
valid_heavy = set(valid_data['seq'])
```

```python
test_data = pd.read_parquet('../data/combined/clustered/final/mode_seq/Galson_2016_single.parquet')

test_data.info()
test_data.head()
```

```python
test_cdr3 = set(test_data['cdr3'])
test_heavy = set(test_data['seq'])
```

```python
training_data = pd.read_parquet('../data/features_data/raw_subject_split/neg_subs/X_train.parquet')

training_data.info()
training_data.head()
```
